```python
import pandas as pd
from openpyxl import load_workbook
from datetime import datetime
```

```python
#이름 입력하도록 설정
path = r"example_file_toyota.xlsm"
```

```python
data = pd.read_excel(path, sheet_name = 0,na_values= ' ') #paht 위치에 있는 엑셀 파일 읽어오기 / na_values= 'NaN'
```

```python
test_data = data
display(test_data.head())

print(test_data.info())
```

```python
#한국어 응답이 하나라도 있는 열 찾아내기
import re
import string

test_data2 = data
pattern = '[ㄱ-ㅎ|ㅏ-ㅣ|가-힣]';
koreacol = ['realid','comp']


complete_data = test_data2[test_data2['comp'] == '{completed}']
for col in complete_data.columns:
    
    complete_data[col] = complete_data[col].apply(str)

    colfilter = complete_data[col].str.contains(pattern, regex = True)
    #print(colfilter)
    if colfilter.any() ==True:
        #print(col + "true")
        koreacol.append(col)

koreancol = complete_data[koreacol]
koreancol.head(5)
```

### 2-2 RecordIP 중복값 표시

```python
#중복있는 값은 True로 체크
test_data['duplicate_chk'] = (test_data.duplicated(['IntroStart.IP']))

#확인
test_data = test_data[test_data['duplicate_chk']==True]
test_data.head()
```

```python
ip_all =list(test_data['IntroStart.IP'])
ip_realid = test_data[['realid','IntroStart.IP']]
# display(ip_realid.head())

ip_unique_list = list((set(ip_all)))

ip_dict = {}
for i in ip_unique_list:
    ip_dict[i] = ip_all.count(i)   

    
ip_dataFrame = pd.DataFrame(list(ip_dict.items()), columns=['IntroStart.IP', 'IP_Count']).sort_values(by='IP_Count',ascending=False)
ip_dataFrame = ip_dataFrame.dropna()
#display(ip_dataFrame.head())


ip_final_table = pd.merge(ip_dataFrame,ip_realid, on='IntroStart.IP', how='left')
ip_final_table
```

### 2-3 COMP 처리

#### 3. comp 행에서 {scr_ 으로 시작하면 전부 {scr_out}으로 치환

```python
test_data.loc[((test_data.comp.str.startswith("{scr_")) | (test_data.comp.str.startswith("{jwquality"))) , "comp"] = "{scr_out}"
test_data
```

```python
#margin의 의미는 모든 값을 합산한 값을 보여주는 인덱스를 생성할 지를 결정하는 인자다.

pivot = pd.pivot_table(test_data, index='comp', values ='realid',aggfunc = 'count', margins = True)
pivotdf = pd.DataFrame(pivot)


pivotdf.loc['LOI(min)'] = test_data['TotalTime2'].median()
pivotdf
```

```python
# 1. 엑셀 파일 열기 w/ExcelWriter
datestring = datetime.strftime(datetime.now(), '%Y%m%d') #저장 날짜로 파일 이름 만들기
filename = 'Respondent_list'+ '_' + datestring.strip() + '_final.xlsx'

# df_needs_columns.to_excel(excel_writer=filename)   
writer = pd.ExcelWriter(filename, engine='xlsxwriter')

# 2. 시트별 데이터 추가하기(1개 시트로 저장할때와 유사)
data.to_excel(writer, sheet_name= 'Raw_Data')
pivotdf.to_excel(writer, sheet_name= 'Pivot_Table')
#test_data.to_excel(writer, sheet_name= 'After_Data')
ip_final_table.to_excel(writer, sheet_name='IP 중복값 Pivot')
koreancol.to_excel(writer,sheet_name='KoreanCol..[update 중]')

# 3. 엑셀 파일 저장하기
writer.save()
```
