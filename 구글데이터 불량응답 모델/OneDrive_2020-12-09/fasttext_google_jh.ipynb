```python
import pandas as pd

df= pd.read_excel('../data/google_data_all_20201210.xlsm')
df.columns =['ID','C1','label']
df['C1'] = df['C1'].apply(str)
df.head()
```

```python
#fasttext에서 사용될 포맷으로 저장하려함
train = open('google_train5','w',encoding='utf-8')
```

```python
#fasttext 포맷으로 저장
#27854
for index,row in df.iterrows():
    #print(row["label"])
    print(f'__label__{row["label"]} {row["C1"]}', file=train)

train.close()
```

```python
#dfTest= pd.read_excel('../data/2019 google BAD sample(1-12).xlsx')

dfTest= pd.read_excel('../data/google_data_all_20201210.xlsm')

dfTest.columns =['ID','C1','label']
dfTest['label_predict'] = ''
dfTest['acc'] = ''

dfTest.head()
```

```python
for index,row in dfTest.iterrows():
    predictArr = model.predict(str(row["C1"]))
    dfTest.at[index,"label_predict"] = int(predictArr[0][0].replace('__label__',''))
    dfTest.at[index,"acc"] = str(predictArr[1][0])
    
```
