```python
#Round2 Data 불러오기
df2 = pd.read_csv('df_유사지역_raw.csv')
df2.head(5)
```

```python
#round_sido_list 에 있는 시로 포함된 애들만 가져오기

df_concat_sido = pd.DataFrame()
for i in round_sido_list:
    df_i = df2[df2['시도명']== i]
    df_concat_sido = pd.concat([df_concat_sido,df_i],axis=0)

#round1_list에 있는 시군구에 포함된 "행정동" 모두 불러오기
df_concat = pd.DataFrame()
for i in round1_list:
    df_i = df_concat_sido[df_concat_sido['시군구명']== i]
    df_concat = pd.concat([df_concat,df_i],axis=0)

print(df_concat['시군구명'].unique())
print(df_concat['행정동명'].unique())
df_concat
```

```python
# #1. round1에서 추려진 시군구가 들어간 행정동 요소를 list로 전부 추출(계산을 위해)
df_concat_list= df_concat.values.tolist()
df_concat_list
```

```python
#정규화하는 함수
#https://jimmy-ai.tistory.com/81
def np_norm(for_norm):
    import numpy as np
    import numpy.linalg as LA

    #for_norm = df_concat.iloc[:,5:].values.tolist()
    df_norm = for_norm/LA.norm(for_norm,ord=2, axis=0)
    return df_norm
```

```python
#정규화한 형태
df_concat_iloc = df_concat.iloc[:,4:]
np_norm(df_concat_iloc)
```

```python
#기준 행정동의 요소들 추출
print(gu,dong)
for row in df_concat_list :
    if (gu in row[2]) and (dong in row[3]) : 
        lista = np.array((row[4:]), dtype=float)

#       정규화 하기 전(비교 차원)
#       print(gu,dong,lista)
        
        #정규화 한 후
        home = np_norm(lista)  # ➊ 사용자로부터 입력받은 지역의 인구 비율을 home에 저장
        print(gu,dong,home)
```

```python
# 그 외 행정동 요소들 추출
for row in df_concat_list :  # ➏ 전체 데이터에 대해 반복
    #print(row)
    away = np_norm(np.array((row[4:]), dtype=float))   # ➋ 그런 다음 다른 지역의 인구 비율을 away에 저장한 후
    #print(away)
    
    #자승제곱법 적용(차이의 제곱의 합)
    s = np.sum((home-away)**2)  # ➌ home에 저장된 값과 away에 저장된 값의 차이의 합을 계산
    
    #자승제곱값을 새로운 컬럼에 추가
    row.append(s)
print(df_concat_list)
```

```python
#3. sum((기준 행정동 요소값 - 각 행정동 값)**2)
df4 = pd.DataFrame(df_concat_list)
df4.columns = ['행정동코드', '시도명', '시군구명', '행정동명', '총인구수', '20대미만', '20대', '30대', '40대',
       '50대', '60대', '70대이상', '남여성비(남)', '남여성비(여)', '연령비_20미만', '연령비_20',
       '연령비_30', '연령비_40', '연령비_50', '연령비_60', '연령비_70이상', '운영가맹점수',
       '가맹점당인구수','차이제곱의합']

#4. 절대값으로 orderby 
df4 = df4.sort_values('차이제곱의합',ascending=True) #key=lambda col: abs(col)
print(len(df4))
df4[['행정동코드','시도명','시군구명','행정동명','차이제곱의합']].head(5)
```
